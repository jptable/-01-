### 資料清理

```python
# 調整資料型態
data['Sex'] = data['Sex'].replace('male', 0)
data['Sex'] = data['Sex'].replace('female', 1)
data['Sex'] = pd.to_numeric(data['Sex'])
```

```python
# change 'Enbarked' with label encoding
#data['Embarked'].value_counts()
data['Embarked'] = data['Embarked'].fillna(0)
data['Embarked'] = data['Embarked'].replace('C', 0)
data['Embarked'] = data['Embarked'].replace('Q', 1)
data['Embarked'] = data['Embarked'].replace('S', 2)
```

```python
# fillna with 1
data['Age'] = data['Age'].fillna(0)
#data['Age']
data['Fare'] = data['Fare'].fillna(0)
data['Fare']
```

```python
# 將不要的欄位另外存放
temp_column = []
temp_column = pd.DataFrame({'Name': data['Name'],
                'Cabin': data['Cabin'],
                'Embarked': data['Embarked'],
                'Ticket': data['Ticket']})
temp_column
```

```python
# 刪掉那些不要的column
data = data.drop(['Name', 'Cabin', 'Ticket'], axis=1)
data
```

### 模型分析

```python
columns_X = set(data.columns) - {'Survived'}
columns_y = ['Survived']
train_X = data[list(columns_X)]
train_y = data[list(columns_y)]

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(),cv=5,scoring='accuracy')
scores
```
